```python
import os
labeled_images = f"{root}/labeled_images"
def remove_with_no_pair():
  a = set(i.split('.')[0] for i in os.listdir(f"{labeled_images}/img"))
  b = set(i.split('.')[0] for i in os.listdir(f"{labeled_images}/lbl"))

  p = a-b
  q = b-a
  c = p.union(q)
  d = []
  for i in os.listdir(f"{labeled_images}/lbl"):
    if i.split('.')[0] in c:
      d.append(f"{labeled_images}/lbl/{i}")

  for i in os.listdir(f"{labeled_images}/img"):
    if i.split('.')[0] in c:
      d.append(f"{labeled_images}/img/{i}")

  for path in d:
    os.remove(path)

remove_with_no_pair()
```

```python
from sklearn.model_selection import train_test_split
image_list = os.listdir(f'{labeled_images}/img')
train_list, test_list = train_test_split(image_list, test_size=0.2, random_state=42)
val_list, test_list = train_test_split(test_list, test_size=0.5, random_state=42)
print('total =',len(image_list))
print('train :',len(train_list))
print('val   :',len(val_list))
print('test  :',len(test_list))
```

```python
from pathlib import Path
from shutil import copyfile

def copy_data(file_list, mode):

    imgs_source = f"{labeled_images}/img"
    img_labels_root = f"{labeled_images}/lbl"

    imgs_dest = f"{labeled_images}/images/{mode}"
    root_file = Path(imgs_dest)
    if not root_file.exists():
        print(f"Path {root_file} does not exit")
        os.makedirs(root_file)

    lbls_dest = f"{labeled_images}/labels/{mode}"
    root_file = Path(lbls_dest)
    if not root_file.exists():
        print(f"Path {root_file} does not exit")
        os.makedirs(root_file)

    for file in file_list:
        img_name = file.replace('.jpg', '')

        img_src_file = f"{imgs_source}/{img_name}.jpg"
        img_dict_file = imgs_dest + '/' + img_name + '.jpg'
        copyfile(img_src_file, img_dict_file)

        label_src_file = f"{img_labels_root}/{img_name}.txt"
        img_dict_file = lbls_dest + '/' + img_name + '.txt'
        copyfile(label_src_file, img_dict_file)
```

```python
copy_data(train_list, "train")
copy_data(val_list, "val")
copy_data(test_list, "test")
```
